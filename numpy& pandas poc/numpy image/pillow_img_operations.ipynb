creating_image_after cleaning the background and recreating using padding and with pixels 

```python
new_img2 = []
for i in range(len(img_mat)):
    if img_mat[i].sum() != 919530:
        recreate= img_mat[i]
        new_pixel = []
        for j in range(len(recreate)):
            if recreate[j].sum() != 765:
#                 print(recreate[j])
                new_pixel.append(recreate[j])
        if new_pixel !=[]:
            new_pixel=np.array(new_pixel,dtype='uint8')
            print(new_pixel)
            new_img2.append(new_pixel)
#             new_img2=np.concatenate(new_img2,new_pixel)
```

```python
print(new_img2)
```

```python
def pad_array_2d(arr, target_shape, padding_value=0):
    padded = np.full(target_shape, padding_value, dtype=arr.dtype)
    padded[:arr.shape[0], :arr.shape[1]] = arr
    return padded

# Find the maximum shape
max_rows = max(arr.shape[0] for arr in new_img2)
max_cols = max(arr.shape[1] for arr in new_img2)
target_shape = (max_rows, max_cols)

# Pad all arrays to the target shape
padded_new_img2 = [pad_array_2d(arr, target_shape) for arr in new_img2]

# Create a numpy array with uint8 type
array = np.array(padded_new_img2, dtype='uint8')
```
