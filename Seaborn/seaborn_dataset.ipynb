```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
flight=sns.load_dataset('flights')
```

```python
flight
```

```python
may_flight = flight.query("month == 'May'")
```

```python
sns.lineplot(data=may_flight,x='year',y='passengers')
```

```python
flight_wide=flight.pivot(index='year',columns='month',values="passengers")
```

```python
flight_wide
```

```python
sns.lineplot(data=flight_wide)
```

```python
tip=sns.load_dataset('tips')
```

```python
tip
```

```python
sns.lineplot(data=tip.head(50),x='tip',y='total_bill',hue='sex',style='sex',dashes=False,markers=True,err_style='bars')
```

```python
#scatter plots
sns.scatterplot(data=tip, x="total_bill", y="tip", hue="time")
```

```python
sns.scatterplot(
    data=tip, x="total_bill", y="tip",hue='size', size="size",
    sizes=(20, 300), hue_norm=(0, 7), legend="full"
)
```

```python
markers = {"Lunch": "s", "Dinner": "X"}
sns.scatterplot(data=tip, x="total_bill", y="tip", style="time", markers=markers)
```

```python
sns.scatterplot(data=tip, x="total_bill", y="tip", hue="time", size="size", style="sex",sizes=(10,100))
```

```python
sns.scatterplot(data=flight_wide)
```

```python
#rel plots also known as relational plots
sns.relplot(data=tip, x="total_bill", y="tip", hue="day", col="time")
```

```python
sns.relplot(data=tip, x="total_bill", y="tip", hue="day", col="time", row="sex")
```

Using kind="line" offers the same flexibility for semantic mappings 

Kind of plot to draw, corresponding to a seaborn relational plot. Options are "scatter" or "line".

```python
sns.relplot(
    data=tip, x="tip", y="total_bill", col="time",
    hue="sex", style="sex", kind='line',markers=True,dashes=False
)
```

A swarm plot is used to show the distribution of data points for different categories while avoiding overlap.

A strip plot also displays the distribution of data points across categories, but it does not avoid overlap.

```python
#strip plot 
from matplotlib import pyplot as plt
fig,ax=plt.subplots(1,2,figsize=(20,7))
sns.stripplot(ax=ax[0],data=tip,y='tip',x='day',hue='sex',dodge=True)
#swam plot
sns.swarmplot(ax=ax[1],data=tip,y='tip',x='day',hue='sex',dodge=True)
```
