In seaborn the graph plotted is more detailed by default we dont need to give lables explicit
In seaborn the the scatter plot can be plotted using diffrent fields from dataset to get better visualization as in below graph:
the x-axis has the tipgiven
the y-axis has the total bill 
the color or hue is set on days like on diffrent day diffrent color marker is used
the size of marker is set on the size attribute from the dataframe, the size in the dataframe represent the size of meal so depending on the size of meal the size of marker is plot

```python
sns.scatterplot(x='tip',y='total_bill',data=tip,hue='day',size='size',palette='YlGnBu')    
```

In histogram we get the labels from sns
the sns histplot provides the parameter kde i.e  kernel density estimation

```python
sns.histplot(tip['tip'],kde=True,bins=20)          
```

```python

# Import libraries
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
 
 
# Creating dataset
x = np.outer(np.linspace(-3, 3, 32), np.ones(32))
y = x.copy().T # transpose
z = (np.sin(x **2) + np.cos(y **2) )
 
# Creating figure
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
 
# Creating plot
ax.plot_surface(x, y, z)
 
# show plot
plt.show()
```
